### pdftotext

Det här är ett litet exempel på hur man i python kan loopa över ett gäng pdf:er och genom pdftotext och regex sedan plocka ut valfri info ur dem. pdftotext måste vara installerat för att detta ska fungera.

```python
# Lägg pdf:er i mappen data som skapas här:
if not os.path.isdir('data'):
    os.makedirs('data')
if not os.path.isdir('data/txt'):
    os.makedirs('data/txt')
if not os.path.isdir('res'):
    os.makedirs('res')
```

```python
# En list comprehension som ger alla pdf:er i mappen data
files = [x for x in os.listdir('data/') if x[-4:] == '.pdf']
```

```python
# Nu en loop över alla filer där 
for file in files:
    filename = file[:-4]
    # Flaggan '-layout' bestämmer att alla dokument ska 
    # behålla sin layout, så tabeller fortfarande står som tabeller:
    os.popen(f'pdftotext -layout -enc UTF-8 data/\
nordenta/{filename}.pdf data/txt/{filename}.txt')
```

```python
def pattern_finder(file):
    res = {}
    res['file'] = file
    with open(f'data/txt/{file}','r') as f:
        text = f.read()
    # bara ett exempel på ett regex som i detta fall hittar personnummer
    p = "[Pp]ersonnummer:\s*(\d+-*\d*)" 
    m = ";".join(re.findall(p,text))
    if not m:
        res['match'] = 'no match'
    else:
        res['match'] = m
    return res
    
    
```

```python
text_files = os.listdir('data/txt')
```

```python
results = []
[results.append(pattern_finder(x)) for x in os.listdir('data/txt')]
```

```python
pd.DataFrame(results).to_excel('res/results.xlsx',index=False)
```
